```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

os.chdir ('C:\\Users\\teas0\\Downloads')
```

```python


results_df_all = pd.read_csv('results_full_feature.csv')
results_df_no_age = pd.read_csv('results_rm_age.csv')
results_df_no_income = pd.read_csv('results_rm_income.csv')
results_df_no_education = pd.read_csv('results_rm_education.csv')
```

```python
dfs = [results_df_all, results_df_no_age, results_df_no_education, results_df_no_income]
model_labels = ['All Features', 'Age Removed Model', 'Education Removed Model', 'Income Removed Model']

for i, df in enumerate(dfs):
    variance_df = df.groupby('Attribute')['FNR'].var().reset_index()
    
    plt.plot(variance_df['Attribute'], variance_df['FNR'], label=model_labels[i])

plt.xlabel('Attributes')
plt.ylabel('Variance')
plt.title('Variance of False Negative Rate (FNR) by Attribute')
plt.legend()
plt.show()
```
