```python
# visualizing the data

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# creating histograms for all the features
feature_data.hist(figsize=(12,25), layout=(7,3))
plt.show()
```

```python
# creating a bar plot for gender

sns.countplot(x='Sex', data=feature_data, palette=['lightcoral','skyblue'])
sex_labels = {0:'Female',1:'Male'}
# renaming the x-axis labels using sex_labels
plt.xticks(ticks=[0,1], labels=['Female','Male'])
plt.show()
```
