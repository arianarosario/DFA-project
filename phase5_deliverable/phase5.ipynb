# 1. Introduction

### a. What are your research questions? What are the inputs, outputs, and evaluation metrics you are interested in, and why?
#### Research Questions
> * Does diabetes prediction accuracy change across income groups?
> * Does diabetes prediction accuracy change across sex groups?
> * Does diabetes prediction accuracy change across age groups?
> * Does diabetes prediction accuracy change across education groups?
> * If there do exist differences, what are the driving factors?

#### Hypothesis
> We hypothesize that diabetes prediction accuracy will change the most prominently across income and education groups, with more minimal change across sex and age groups. We believe that by removing income, education, and age subgroups in model training to account for an unbalanced dataset, we can generate a fairer model through optimizing for predictive equality in equal False Negative Rates across demographic subgroups.

#### Inputs
> * Our inputs include: HighBP, HighChol, CholCheck, BMI, Smoker, Stroke, HeartDiseaseorAttack, PhysActivity, Fruits, Veggies, HvyAlcoholConsump, AnyHealthcare, NoDocbcCost, GenHlth, MentHlth, PhysHlth, and DiffWalk.
> * The demographic input data are Sex, Age, Education, and Income.

#### Outputs
> Our output is the binary prediction such that 1 is the patient is diagnosed with diabetes or prediabetes and 0 represents that the patient is not diabetic.

#### Evaluation metrics
> The evaluation metrics we want to use for our algorithm include assessing if the F1 socre (precision, recall), and FNR is different for various incom, age, education, and sexx groups.


### b. Why is this an interesting / important research question? How does it relate to algorithmic fairness?
> The National Diabetes Statistics Report [1] reveals that out of 38.4 million Americans living with diabetes, 29.7 million are diagnosed, leaving 8.7 million unaware of their condition. This gap in diagnosis highlights a critical area for investigation, especially regarding the role of predictive algorithms in healthcare and their potential biases towards certain demographics. We are focusing our analysis on gender and income disparities, as these factors have shown significant variations in the prevalence and awareness of prediabetes and diabetes. The rationale behind our focus is the potential for biases in machine learning algorithms, which might contribute to the higher rates of undiagnosed diabetes among specific groups. Understanding and addressing these biases is important because early and accurate diagnosis of diabetes is essential for preventing serious complications, such as heart disease, vision loss, and kidney disease. Our research is driven by the goal of enhancing diagnostic accuracy and ensuring equitable health outcomes for all individuals, regardless of gender, age, income, and education levels.

# 5. Modeling

#### Functions

```python
def split_data(X, y):
    '''Load and split data into train:validation:test = 6:2:2
    Args:
        - X: pandas DataFrame, features only
        - y: pandas DataFrame, targets only
    Return:
        (X_train, X_val, X_test), (y_train, y_val, y_test)
    '''
    # X, y = features, target.Diabetes_binary
    X_train_test, X_test, y_train_test, y_test = train_test_split(X, y, test_size=0.2)
    X_train, X_val, y_train, y_val = train_test_split(X_train_test, y_train_test, test_size=0.25)
    return (X_train, X_val, X_test), (y_train, y_val, y_test)
```

```python
def train(X_train, X_val, y_train, y_val, params, exclude_features):
    '''Train a model with give param and features to exclude
    Args:
        - params: dict, model parameters
        - exclude_features: set, feature name to exclude
    Return:
        lgb model
    '''
    # Exclude features
    feat = [i for i in list(X_train.columns) if i not in exclude_features]
    # Build Data
    lgb_train = lgb.Dataset(X_train[feat], y_train, free_raw_data=False)
    lgb_val = lgb.Dataset(X_val[feat], y_val, reference=lgb_train, free_raw_data=False)
    # Build Model
    evals_result = {}
    gbm = lgb.train(params, # parameter dict to use
                    lgb_train,
                    num_boost_round=10000, # the boosting rounds or number of iterations.
                    early_stopping_rounds=50, # early stopping iterations.
                    # stop training if *no* metric improves on *any* validation data.
                    valid_sets=[lgb_train, lgb_val],
                    evals_result=evals_result, # dict to store evaluation results in.
                    verbose_eval=100) # print evaluations during training.
    
    y_val_pred = gbm.predict(X_val[feat], num_iteration=gbm.best_iteration)
    auc_score = roc_auc_score(y_val, y_val_pred)
    print('Validation set evaluation:')
    print(f'AUC score: {round(auc_score, 4)}')
    # find best threshold through validation set
    precision, recall, thresholds = precision_recall_curve(y_val, y_val_pred)
    f1_scores = 2*recall*precision/(recall+precision)
    threshold = thresholds[np.argmax(f1_scores)]
    y_val_pred = np.where(y_val_pred > threshold, 1, 0)
    f1 = f1_score(y_val, y_val_pred)
    print(f"F1-Score: {round(f1, 4)}    (Best threshold: {round(thresholds[np.argmax(f1_scores)], 4)})")
    
    # Calculating FNR
    conf_matrix = confusion_matrix(y_val, y_val_pred)
    FN, TP = conf_matrix[1][0], conf_matrix[1][1]
    FNR = FN / (FN + TP)
    print(f'FNR: {round(FNR, 4)}')
    
    return gbm, round(threshold, 2), feat
```

```python
def cm_plot(y_test, y_test_pred, title):
    cm = confusion_matrix(y_test, y_test_pred, labels=[0, 1])
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['negative', 'positive'])
    disp = disp.plot(cmap=plt.cm.Blues,values_format='g')
    # disp.plot()
    plt.title(f'LightGBM ({title})')
    plt.show()
    # calculate score
    score = f1_score(y_test, y_test_pred)
```

```python
def predict(model, data, target, feat, threshold, title):
    # Run model
    pred_prob = model.predict(data[feat], num_iteration=model.best_iteration)
    # print evaluation metrics
    auc_score = roc_auc_score(target, pred_prob)
    print(f'AUC score: {round(auc_score, 4)}')
    pred = np.where(pred_prob > threshold, 1, 0)
    f1 = f1_score(target, pred)
    print(f"F1-Score: {round(f1, 4)}")
    conf_matrix = confusion_matrix(target, pred)
    FN, TP = conf_matrix[1][0], conf_matrix[1][1]
    FNR = FN / (FN + TP)
    print(f'FNR: {round(FNR, 4)}')
    cm_plot(target, pred, title)
    # return eval dataset by subgroups
    sensitive_attributes = ['Sex', 'Age', 'Education', 'Income']
    AUC_subgroups = {}
    FNR_subgroups = {}
    FPR_subgroups = {}

    for att in sensitive_attributes:
        AUC_subgroups[att] = {}
        FNR_subgroups[att] = {}
        FPR_subgroups[att] = {}

        subgroups = features[att].unique()
        for subgroup in subgroups:
            curr_subgroup_index = data[att] == subgroup
            curr_y_test = target[curr_subgroup_index]
            curr_y_pred_proba = pred_prob[curr_subgroup_index]
            curr_y_pred = pred[curr_subgroup_index]
            curr_AUC = roc_auc_score(curr_y_test, curr_y_pred_proba)
            tn, fp, fn, tp = confusion_matrix(curr_y_test, curr_y_pred).ravel()
            curr_FNR = fn/(tp+fn)
            curr_FPR = fp/(tn+fp)

            AUC_subgroups[att][subgroup] = curr_AUC
            FNR_subgroups[att][subgroup] = curr_FNR
            FPR_subgroups[att][subgroup] = curr_FPR
    temp_data = [(att, subgroup) for att, subgroups in AUC_subgroups.items() for subgroup in subgroups]
    results_df = pd.DataFrame(temp_data, columns=['Attribute', 'Subgroup'])
    results_df['AUC'] = [AUC_subgroups[att][subgroup] for att, subgroups in AUC_subgroups.items() for subgroup in subgroups]
    results_df['FNR'] = [FNR_subgroups[att][subgroup] for att, subgroups in FNR_subgroups.items() for subgroup in subgroups]
    results_df['FPR'] = [FPR_subgroups[att][subgroup] for att, subgroups in FPR_subgroups.items() for subgroup in subgroups]
    results_df.sort_values(by=['Attribute', 'Subgroup'], inplace=True)
    
    return results_df
```

### 6. a Interpreting Results

#### AUCs
Across all our models, we're getting decent AUC scores and we can see they're also evenly distributed across demographic subgroups. Meaning that for regardless of the patient's age, sex, education or income level, the model is able to predict diabetes with similar accuracy. This could be an indication that the model is performing fairly but we run into issues when we look into FNRs and FPRs across subgroups. 

#### False Negatives and False Positives
Turns out that the model gives false negatives and false positives at different rates for different subgroups. This is a problem because it means that the model is not equally accurate for all subgroups. We should *specially pay attention to False Negatives since they stand to cause the most harm* (i.e. incorrectly predicting a patient is healthy when they're not). And this imbalance is specially evident across age groups and income levels with **some groups getting unacceptably high False Negative rates.**

### 6.b Hypothesis

In our hypothesis, we theorized that maybe removing some of the most unbalanced features (such as Age, Income and Education) could maybe help alleviate disparities in FNRs and FPRs. To do this, we testing re-training our model with some of these unbalanced features removed. 

However, we found that method to have little success. For instance, when we removed Age, the model had slightly lower FNRs across Age subgroups but worse FNR disparities across other groups. On the other hand, when we removed Income, the FNRs improved across both Education and Income groups, with Age remaining high. 

With this in mind, we think we can **reject our hypothesis** and conclude that removing these features does not help alleviate disparities in FNRs and FPRs across demographic subgroups.

# 7. Conclusions and Limitations

### 7.a Hypothesis and Analysis

#### Our Model
As mentioned above, we came up with a decently performing Tree-based model to predict diabetes. When evaluating it, we found that it performed decently with an average AUC score of over 0.8. However, we found that this accuracy was unfairly distributed across demographic groups with the model performing worse for certain groups. Specifically, we found unacceptably high False Negative Rates for certain age and income groups.

#### Our Hypothesis
With this disparities in model performance, we theorized that maybe removing some of these unbalanced features could improve performance. So, we re-trained 3 more versions of the model, each with one of the sensitive features removed (age, education, income). 

However, we found that *this method did not help alleviate disparities in FNRs and FPRs* So, **we reject the hypothesis** and conclude that removing these features does not lead to a better model performance.

With this in mind, we have to go back to the drawing board and think of other ways to improve performance while ensuring fair outcomes across demographic groups.

### 7.b Limitations

#### Data
If we look at the group distribution (visualized in part 6), we can see considerable imbalances across demographic features with **some groups being considerable unrepresented**. For instance, groups with lower education and income levels represent a very small fraction of the data (e.g. less than  1.7% of people with education levels 1 or 2)

#### Model
We used a tree-based (LightGBM) model to predict diabetes. While the model has somewhat decent general AUC scores, we noticed that the False Positives and False Negative were not evenly distributed and disproportionally affected some groups. 

#### Analysis
Our fairness analysis is mainly based on FPR and FNR balance. We assume that False Negatives stand to cause the most harm to patients given that they'd be incorrectly classified as healthy. So, with this in mind, we hope that a fair model would have minimal and balance FNRs.  However, we did **not consider other fairness metrics** such as Equalized Odds or Demographic Parity which could have given us a more nuanced view of fairness in our model.
